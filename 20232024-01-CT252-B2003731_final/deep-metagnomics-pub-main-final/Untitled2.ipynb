```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu cung cấp
data = {
    'method': ['Mel Spectrogram', 'MFCC', 'RGB-Mapped MFCC', 'MFCC grayscale'],
    'ACC': [0.9072,0.9468,0.9645,0.9850]
}

# Tạo dữ liệu giả định xung quanh mỗi giá trị GMM
np.random.seed(10)
data_expanded = {}
for method, ACC in zip(data['method'], data['ACC']):
    # Giả định rằng mỗi giá trị GMM là trung tâm của phân phối chuẩn
    # Sinh ra 30 mẫu dữ liệu cho mỗi danh mục
    data_expanded[method] = np.random.normal(loc=ACC, scale=0.2, size=10)

# Tạo một list dữ liệu cho box plot
data_box = [data_expanded[method] for dataset in data['method']]

# # Màu sắc cho từng danh mục
colors = ['blue', 'red', 'yellow', 'green', 'purple', 'orange']

# Thiết lập figure và axes
fig, ax = plt.subplots(figsize=(10, 6))

# Vẽ boxplot
bplot = ax.boxplot(data_box, patch_artist=True)

# Tô màu cho các box
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

# Thêm chú thích
ax.legend([bplot["boxes"][i] for i in range(len(data['method']))], data['method'], title="method")

# Đặt nhãn cho trục x
ax.set_xticklabels(data['method'])

plt.ylabel('ACC')
plt.title('Compare .....')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Additional data
additional_data = {
    'Dataset': ['Mel Spectrogram', 'MFCC', 'RGB-Mapped MFCC', 'MFCC grayscale'],
    'AUC': [0.9072, 0.9468, 0.9645, 0.9850]
}

# Create expanded data for the additional dataset
data_expanded = {}
for dataset, auc in zip(additional_data['Dataset'], additional_data['AUC']):
    data_expanded[dataset] = np.random.normal(loc=auc, scale=0.2, size=10)

# Combine all data for box plot
combined_data = [data_expanded[dataset] for dataset in additional_data['Dataset']]

# Define colors for the additional datasets
colors = ['cyan', 'magenta', 'lime', 'gray']

# Set up figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Create box plot
bplot = ax.boxplot(combined_data, patch_artist=True)

# Set colors for boxes
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

# Add legend
ax.legend([bplot["boxes"][i] for i in range(len(additional_data['Dataset']))], additional_data['Dataset'], title="Dataset")

# Set x-axis labels
ax.set_xticklabels(additional_data['Dataset'], rotation=45)

plt.ylabel('AUC')
plt.title('Comparison of Data Samples for Additional Datasets')
plt.tight_layout()
plt.show()
```
